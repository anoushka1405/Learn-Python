```python
#1 Fibonacci Number

def isFibonacci(n):
    a,b = 0,1

    if n==0 or n==1:
        return True
    
    while(b<n):
        a,b = b,a+b
    
    return b==n

n1 = int(input("Enter Number : "))
print(isFibonacci(n1))
```

```python
#2. Printing nth multiple of a number k in the Fibonacci Series

def nthFibonacci(n,k):
    a,b = 0,1
    count = 0

    while count!=n:
        a,b = b,a+b
        if b%k==0:
            count+=1
    
    return b

n2 = int(input("Enter n:"))
k = int(input("Enter k:"))

print(nthFibonacci(n2,k))
```

```python
#4. Sum of squares of first n natural numbers
def sumSq(n):
    count = 1
    sum = 0

    while count!=n+1:
        sum+=count*count
        count+=1
    return sum

n = int(input("Enter n:"))
print(sumSq(n))
```

```python
#5. Sum of cubes of first n natural numbers

def cubeSum(n):
    sum = 0
    for i in range(1,n+1):
        sum+=i*i*i
    return sum

n = int(input("Enter n : "))
print(cubeSum(n))
```

```python
#6. Array Rotation - Reversal Algorithm

def reverse(arr, start, end):
    while start<end:
        arr[start],arr[end] = arr[end],arr[start]
        start+=1
        end-=1

def rotate(arr, d):
    n = len(arr)
    d = d%n

    reverse(arr,0,d-1)
    reverse(arr,d,n-1)
    reverse(arr,0,n-1)

    return arr

arr = list(map(int,input("Enter array elements : ").replace(',','').split()))
d = int(input("Enter number of rotations : "))
```

```python
#9. To check is array is monotonic

def isMonotonic(arr) : 
    inc = True
    dec = True

    for i in range(1,len(arr)) : 
        if arr[i]>arr[i-1]:
            dec = False
        if arr[i]<arr[i-1]:
            inc = False
    
    return inc or dec

arr = list(map(int,input("Enter array elements : ").replace(',','').split()))

if(isMonotonic(arr)):
    print("array is monotonic")
else:
    print("not monotonic")
```

```python
#11. Program to Remove multiple elements from a list

def removeMultiple(lst,remove_list):
    remove_set = set(remove_list)
    return [x for x in lst if x not in remove_set]

lst = list(map(int, input("Enter : ").replace(',', ' ').split()))
remove_list = list(map(int, input("Enter elements to remove : ").replace(',', ' ').split()))

print("Resultant List : ",removeMultiple(lst, remove_list))
```

```python
#Program 12 - Remove empty list from list

def removeEmpty(lst):
    return[x for x in lst if x!=[]]

lst = [[1,2],[],[3,4],[],[],[5]]

print("original list : ",lst)
print("After removing : ",removeEmpty(lst))
```

```python
# Remove empty tuples from list

def removeEmptyTuples(lst):
    return [t for t in lst if t != ()]

# Example list
lst = [(1, 2), (), (3, 4), (), (), (5,)]

print("Original list:", lst)
print("After removing empty tuples:", removeEmptyTuples(lst))
```
