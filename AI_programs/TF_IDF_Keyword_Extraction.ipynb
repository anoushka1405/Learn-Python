To extract important keywords from text documents using TF-IDF (Term Frequency–Inverse Document Frequency) 
with Python’s Scikit-Learn library.


TF-IDF (Term Frequency–Inverse Document Frequency) is a technique used in Natural Language Processing (NLP) to measure the importance of a word in a document relative to a collection of documents.

Term Frequency (TF) refers to how frequently a word appears in a document. Inverse Document Frequency (IDF) measures how important a word is by reducing the weight of commonly used words and increasing the weight of rare words.
TF-IDF is calculated as: TF-IDF = TF × IDF

Words that appear frequently in a document but rarely in others get higher scores, while common words like “the”, “is”, etc., get lower scores.
In this experiment, Scikit-Learn’s TfidfVectorizer is used to convert text into numerical form and compute TF-IDF values to identify important keywords.


```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#Sample dataset
data = {'documents': ['The quick brown fox jumps over the lazy dog.',
                      'Never jump over the lazy dog quickly.',
                      'Bright vixens jump; dozy fowl quack.',
                      'Quick brown dogs jump over the lazy fox.']}
```

```python
df = pd.DataFrame(data)
```

```python
#Compute TF-IDF
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df['documents'])
```

```python
#Extract keywords
feature_names = vectorizer.get_feature_names_out()
dense = tfidf_matrix.todense()
denselist = dense.tolist()

df_tfidf = pd.DataFrame(denselist, columns = feature_names)
```

```python
#Display the TF-IDF scores
print(df_tfidf)
```

The TF-IDF technique was successfully implemented using Scikit-Learn. 
Important keywords were extracted from the text dataset, and their TF-IDF scores were generated. 
Words with higher values indicate greater importance in their respective documents.
