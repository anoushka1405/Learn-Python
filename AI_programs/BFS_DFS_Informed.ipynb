Breadth-First Search (BFS):

- BFS is an uninformed search algorithm that explores nodes level by level.
- It uses a queue (FIFO) to store nodes to visit.
- BFS guarantees the shortest path in terms of the number of edges (for unweighted graphs).
- Applications: Shortest path, web crawling, peer-to-peer networks.

```python
from collections import deque
#Sample Graph as adjacency list

graph = {
    'A' : ['B','C'], 
    'B' : ['D','E'],
    'C' : ['F'],
    'D' : [],
    'E' : ['F'],
    'F' : []
}

def bfs(start, goal) : 
    visited = set()
    queue = deque([start])

    while(queue):
        node = queue.popleft()
        print("Visiting : ",node)

        if(node==goal):
            print("Goal Found")
            return True
        
        if node not in visited:
            visited.add(node)
            queue.extend(graph[node])
        
    return False

print(bfs('A','F'))
```

Depth-First Search (DFS):

- DFS is an uninformed search algorithm that explores nodes deeply along one path before backtracking.
- It uses a stack (LIFO) or recursion.
- DFS does not guarantee the shortest path but uses less memory than BFS.
- Applications: Topological sorting, cycle detection, puzzle solving.

```python
def dfs(start,goal):
    visited = set()
    stack = [start]

    while stack : 
        node = stack.pop()
        print("Visiting : ",node)
        if node==goal:
            print("Goal Found")
            return True
        if node not in visited:
            visited.add(node)
            stack.extend(graph[node])
    return False

print(dfs('A','F'))
```
